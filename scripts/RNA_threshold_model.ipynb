```python
import numpy as np
from matplotlib import rc
from matplotlib import pyplot as plt
```

```python
def threshold_arctan(x, t, r):
    return r * (0.5 + (1 / np.pi) * np.arctan(1e2 * (x - t)))

n = 5000
ivl = 5
x = np.linspace(0, 11, n)
t = 5
r = 0.9
c = int(n/2)

thresh_arctan = threshold_arctan(x, t, r)


dots = np.random.uniform(0, 1, 300)
dotspos = np.random.beta(2, 1, 300) * 11

cm = 1/2.56  # convert cm to inch

rc("font", family='monospace', size=16)
rc("font", family='sans-serif', size=16)

fig, ax = plt.subplots(1, 1, figsize=(6*cm,4.5*cm))
ax.scatter(dotspos, dots, marker="o", s=7.5, color="#B56478", alpha=.5)
ax.vlines(t, -0.05, 1.0, lw=.5, ls="--", color="black")
ax.plot(x, thresh_arctan, lw=2, ls="-", color="#5A2731", label="threshold arctan")
ax.set_yticks([])
ax.set_xticks([])
ax.set_xticklabels([])
# ax.set_ylabel("NRF2 expression rate")
ax.set_xlim(0, 11)
ax.set_ylim(-0.05, 1.1)
ax.set_xlabel(r"Concentration ($C_i$)")
ax.text(11.2, r, r"$r_{rt}$", va="center", ha="left")
ax.text(-0.2, 0, r"$0$", va="center", ha="right")
ax.text(t, 1.01, r"$z_{c_i}$", va="bottom", ha="center")

ax.spines[["top", "left", "right"]].set_visible(False)
fig.subplots_adjust(bottom=0.2, right=0.85)
fig.savefig("../../../results/plots/activation.svg", transparent=True)
```
